Cleaning the data

```python
df.isnull().sum
```

```python
df=df.drop(['approxdate'],axis=1)
```

```python
df.head()
```

```python
df.rename(columns={"imonth":"Month","region_txt":"Region","country_txt":"Country"},inplace= True)
```

```python
df.head()
```

```python
df["nkill"]=df["nkill"].fillna(0)
df["nwound"]=df["nwound"].fillna(0)
df["Casualty"]=df["nkill"]+df["nkill"]
```

```python
df.describe()
```

Visualizing the data

Number of Attacks in each year

```python
attacks=df["iyear"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"iyear","iyear":"Attacks"}).set_index("iyear")
attacks.head()
```

```python
attacks.plot(kind="bar",color="cornflowerblue",figsize=(15,6),fontsize=13)
plt.title("Timeline of Attacks",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

Killed in each year

```python
killed=df[["iyear","nkill"]].groupby("iyear").sum()
killed.head()
```

Wounded in each year



```python
wounded=df[["iyear","nwound"]].groupby("iyear").sum()
wounded.head()
```

```python
print("Country with the most attacks:",df['Country'].value_counts().idxmax())
print("City with the most attacks:",df['city'].value_counts().index[1])
print("Year with the most attacks:",df['iyear'].value_counts().idxmax())
print("Month with the most attacks:",df['Month'].value_counts().idxmax())
print("Most Attack Types:",df['attacktype1_txt'].value_counts().idxmax())
```

Number of attacks over the year

```python
print("iyear   Number of Attacks")
df['iyear'].value_counts(dropna = False).sort_index()
```

```python
x_year = df['iyear'].unique()
y_yearly_attacks = df['iyear'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (20,10))
sns.barplot(x = x_year,
           y = y_yearly_attacks,
           palette = 'rocket')
plt.xlabel('Year of Attacks')
plt.ylabel('Number of Attacks each year')
plt.title('Attack by Years')
plt.xticks(rotation=60)
plt.show()
```

Terrorist activities by region in each year through area plot

```python
pd.crosstab(df.iyear, df.Region).plot(kind='area',figsize=(14,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

Region-wise attacks

Distribution of terrorist attacks over regions from 1970-2017

```python
reg=pd.crosstab(df.iyear,df.Region)
reg.head()
```

```python
reg.plot(kind="area", stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```
